### Importing relevant libraries and modules


```python

import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import datetime as dt
```

### Loading the dataset into a DataFrame


```python

df = pd.read_csv(r"E:\Users\shipr\Downloads\Files (2)\Files\Unicorn_Companies.csv")
```

```python
# Displaying the first 5 rows of the data.

df.head()
```

```python
# Checking for duplicates.

df.drop_duplicates()
df.shape
```

```python
# Displaying the data types of the columns.

df.info()
```

```python
# Sorting companies and displaying the first 20 rows of the resulting DataFrame.

df.sort_values(by='Year Founded',ascending=True).head(20)
```

```python
# Displaying each unique year that occurs in the dataset
# along with the number of companies that were founded in each year.

df.groupby('Year Founded').agg(count_of_companies=('Company','count'))
```

### Converting the `Date Joined` column to datetime


```python
# Converting the `Date Joined` column to datetime.

df['date_joined']=pd.to_datetime(df['Date Joined'])
df.head(10)
```

### Creating a `Month Joined` column

```python
# creating a Month Joined column.

df['month_joined']=df['date_joined'].dt.strftime('%B')
df.head()
```

### Creating a `Years To Join` column


```python
# Determining how many years it took for companies to reach unicorn status to create a `Years To Join` column.

df['years_to_join']=df['date_joined'].dt.year-df['Year Founded']
df.head()
```

### Gaining more insight on year 2021


```python
# Filtering dataset by year 2021

df2021= df[df['date_joined'].dt.year==2021].copy()
df2021.head()
```

### Observing trends over time

```python
# Creating a week joined column and grouping by the year 2021 to observe trends

df2021['week_joined']= df2021['date_joined'].dt.strftime('%Y-W%V')
df2021.head(10)
df2021byweek=df2021.groupby('week_joined').agg(count_of_companies=('Company','count')).reset_index()
df2021byweek
```

### Comparing trends over time

```python

# Filtering by the year 2020 and concatenating that new subset with the subset of year 2021
# Adding a new column `quarter`.

df2020= df[df['date_joined'].dt.year==2020].copy()
df2020.head()
uniondf=pd.concat([df2020,df2021.drop(columns='week_joined')])
uniondf.tail()
uniondf['quarter']=uniondf['date_joined'].dt.to_period('Q').dt.strftime('%Y-Q%q')
uniondf.head()
```

```python
# Converting the Valuation column to numeric

uniondf['Valuation']=uniondf['Valuation'].str.strip('$B').astype(float)
```

```python
uniondf.head()
```

```python
# Grouping valuation by quarter

df3=uniondf.groupby('quarter')['Valuation'].mean().reset_index()
df3.head()
```

### Visualizing the time it took companies to become unicorns

Creating a box plot to visualize the distribution of how long it took companies to become unicorns, with respect to the month they joined. 

```python
# Defining a list that contains months in chronological order.

month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", 
               "December"]

print(month_order)
          
```

```python
# Creating a box plot to visualize the distribution of how long it took companies to become unicorns, with respect to the month they joined.

sns.boxplot(data=df,x='month_joined',y='years_to_join',order=month_order,showfliers=False)
plt.plot

plt.title('Years to become Unicorn by month joined')

plt.xticks(rotation=45, horizontalalignment='right')
 
plt.show()



```

### Visualizing the time it took companies to reach unicorn status
Creating a bar plot to visualize the average number of years it took companies to reach unicorn status with respect to when they were founded. 

```python
# Setting the size of the plot.
plt.figure(figsize=(10,6))



sns.barplot(x=df["Year Founded"], y=df["years_to_join"],errorbar=('ci', False))
plt.plot()

plt.title('Time to become Unicorn by year joined')
plt.xticks(rotation=45,horizontalalignment='right')
plt.show()
```

### Visualizing the number of companies that joined per week

Using the subset of companies joined in 2021, grouped by the week, creating a bar plot to visualize the number of companies that joined per week for that year. 

```python

plt.figure(figsize=(20,5))

sns.barplot(data=df2021byweek, x='week_joined', y='count_of_companies')
plt.plot()

plt.title('Number of companies to become Unicorn per week in 2021')
plt.xticks(rotation=45,horizontalalignment='right')
plt.show()

```

### Visualizing the average valuation over the quarters

Using the subset of companies that joined in 2020 and 2021, creating a grouped bar plot to visualize the average valuation over the quarters.

```python
# Using slicing, extracting the year and the quarter and adding two new columns into the subs

df3['Quarter']=df3.quarter.str[-2:]

df3['Year']=df3.quarter.str[:4]
df3.head()

# Plotting the chart

plt.figure(figsize=(10,5))

sns.barplot(data=df3, x='Quarter', y='Valuation',hue='Year')
plt.plot()

plt.title('Average valuation of quarters for the years 2020 and 2021')
plt.show()
```
